# Tipos de Dados no Python

No Python, os tipos de dados representam as diferentes naturezas das informações com as quais podemos trabalhar, como números, textos, valores booleanos e estruturas mais complexas. Compreender os tipos de dados é essencial para desenvolver códigos corretos e eficientes, pois cada tipo tem seu comportamento específico e métodos associados. Abaixo, apresentamos exemplos práticos dos principais tipos de dados em Python, incluindo inteiros, ponto flutuante, strings, booleanos, listas, tuplas, dicionários e conjuntos.

```python
# Tipo inteiro (int)
idade = 30
print("Tipo de idade:", type(idade), "| Valor:", idade)
```

```python
# Tipo ponto flutuante (float)
altura = 1.75
print("Tipo de altura:", type(altura), "| Valor:", altura)
```

```python
# Tipo string (str)
nome = "Ana"
print("Tipo de nome:", type(nome), "| Valor:", nome)
```

```python
# Tipo booleano (bool)
tem_carteira = True
print("Tipo de tem_carteira:", type(tem_carteira), "| Valor:", tem_carteira)
```

```python
# Tipo lista (list)
compras = ["arroz", "feijão", "ovo"]
print("Tipo de compras:", type(compras), "| Valor:", compras)
```

```python
# Tipo tupla (tuple)
coordenadas = (10.5, 20.3)
print("Tipo de coordenadas:", type(coordenadas), "| Valor:", coordenadas)
```

```python
# Tipo dicionário (dict)
aluno = {"nome": "João", "idade": 22, "curso": "Python"}
print("Tipo de aluno:", type(aluno), "| Valor:", aluno)
```

```python
# Tipo conjunto (set)
numeros = {1, 2, 3, 2, 1}
print("Tipo de numeros:", type(numeros), "| Valor:", numeros)
```

```python
# Tipo None (ausência de valor)
resultado = None
print("Tipo de resultado:", type(resultado), "| Valor:", resultado)
```
