# Estruturas no Python

As estruturas de controle e repetição são fundamentais para criar fluxos de execução inteligentes e adaptáveis em Python. Elas permitem tomar decisões com base em condições (como o if, elif e else) e repetir blocos de código automaticamente (for, while). Esses recursos tornam possível automatizar tarefas, validar entradas e iterar sobre coleções de dados. A seguir, mostramos exemplos práticos dessas estruturas e suas variações mais comuns.

```python
# Estrutura condicional simples

idade = 18

if idade >= 18:
    print("Você é maior de idade.")
```

```python
# Estrutura condicional composta

if idade < 18:
    print("Menor de idade.")
else:
    print("Maior de idade.")
```

```python
# Estrutura condicional com elif e else

nota = 7.5
if nota >= 9:
    print("Excelente!")
elif nota >= 7:
    print("Bom!")
elif nota >= 5:
    print("Regular.")
else:
    print("Reprovado.")
```

```python
# Operador ternário

mensagem = "Aprovado" if nota >= 7 else "Reprovado"
print("Resultado:", mensagem)
```

```python
# Estrutura de repetição FOR com Range

# for com range
print("Contando de 1 a 5:")
for i in range(1, 6):
    print(i)
```

```python
# Estrutura de repetição FOR com Lista

nomes = ["Ana", "Bruno", "Carlos"]
print("Lista de nomes:")
for nome in nomes:
    print(nome)
```

```python
# Estrutura de repetição FOR com Enumerate

for i, nome in enumerate(nomes):
    print(f"Posição {i}: {nome}")
```

```python
# for com break e continue
print("Usando break e continue:")
for i in range(10):
    if i == 5:
        break  # Para quando i == 5
    if i % 2 == 0:
        continue  # Pula os pares
    print(i)  # Imprime só os ímpares menores que 5
```

```python
# Estrutura de repetição WHILE

contador = 0
print("Contador com while:")
while contador < 5:
    print("Valor:", contador)
    contador += 1
```

```python
# Estrutura de repetição WHILE com break
numero = 0
while True:
    print("Número:", numero)
    numero += 1
    if numero > 3:
        break  # Sai do loop ao atingir 4
```
